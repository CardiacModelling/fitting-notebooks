# Appendix K: Compensated patch-clamp models

In [Appendix F](./appendix-E-non-ideal-op-amp.ipynb) we compared uncompensated patch-clamp models, omitting voltage offset and leak current for simplicity.
In this appendix, we compare models of _compensated_ patch clamp, again omitting offset and leak.

The schematic is shown below:

<img src="resources/patch-comp-6-no-leak-eoff.png" style="margin:auto" />

## Sigworth-style model (1, 2a, 3a, 4, 5a)

\begin{align}
1. && C_m\dot{V}_m = \frac{V_p - V_m}{R_s} - I
\end{align}

\begin{align}
2a. && (C_p + C_f)\dot{V}_p = \frac{V_o-V_p}{R_f} + C_f\dot{V}_o - \frac{V_p-V_m}{R_s} + (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

\begin{align}
3a. && \tau_a\dot{V}_o = V_\text{ref} - V_p
\end{align}

\begin{align}
4. && \tau_\text{sum} \dot{V}_\text{ref} = V_c + \alpha R_s^* I_\text{obs} - V_\text{ref}
\end{align}

\begin{align}
5a. && R_f I_\text{obs} = V_o - V_\text{ref}
\end{align}

## Alternative op-amp equation (1, 2b, 3b, 4, 5a)

\begin{align}
1. && C_m\dot{V}_m = \frac{V_p - V_m}{R_s} - I
\end{align}

\begin{align}
2b. && C_f\dot{V}_o = \frac{V_p-V_o}{R_f} + \left(C_p+C_f\right)\dot{V}_p + \frac{V_p-V_m}{R_s} - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

\begin{align}
3b. && \tau_c\dot{V}_p = V_\text{ref} - V_p
\end{align}

\begin{align}
4. && \tau_\text{sum} \dot{V}_\text{ref} = V_c + \alpha R_s^* I_\text{obs} - V_\text{ref}
\end{align}

\begin{align}
5a. && R_f I_\text{obs} = V_o - V_\text{ref}
\end{align}

## Lei-style model (1, 2c, 3b, 4, 5a) or (1, 2b, 3b, 4, 5b)

Following Lei et al. (2020), we get

\begin{align}
1. && C_m\dot{V}_m = \frac{V_p - V_m}{R_s} - I && \text{(Equation 2.3)}
\end{align}

\begin{align}
3b. && \tau_c\dot{V}_p = V_c - V_p && \text{(Equation 2.4)}
\end{align}

\begin{align}
4. && \tau_\text{sum} \dot{V}_\text{ref} = V_c + \alpha R_s^* I_\text{obs} - V_\text{ref} && \text{(Equation 2.5)}
\end{align}

and we can combine two equations to find

\begin{align}
I_\text{in} &= I + C_p \dot{V}_p + C_m \dot{V}_m - (C_p^* + C_m^*) \dot{V}_\text{ref} && \text{(Equation 2.6)} \\
            &= \frac{V_p - V_m}{R_s} + C_p \dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref} && \text{(Equation 2.3)}
\end{align}

Then, we define

\begin{align}
5a. && R_f I_\text{obs} = V_o - V_\text{ref} && \text{(No equivalent)}
\end{align}

and insert into Lei et al. Equation 2.7 to find

\begin{align}
R_f C_f \dot{I}_\text{obs} &= I_\text{in} - I_\text{obs} && \text{(Equation 2.7)} \\
C_f (\dot{V}_o - \dot{V}_\text{ref}) &= \frac{V_p - V_m}{R_s} + C_p \dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref} - \frac{V_o - V_\text{ref}}{R_f} &&
\end{align}

\begin{align}
2c. && C_f \dot{V}_o &= \frac{V_\text{ref} - V_o}{R_f} + C_p \dot{V}_p + C_f \dot{V}_\text{ref} + \frac{V_p - V_m}{R_s} - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

For a model (1, 2c, 3b, 4, 5a).

**Alternatively**, and equivalently, we define

\begin{align}
5b. && R_f I_\text{obs} = V_o - V_p && \text{(No equivalent)}
\end{align}

with which we can derive 2b. from Lei et al. Equation 2.7:

\begin{align}
C_f (\dot{V}_o - \dot{V}_p) &= \frac{V_p - V_m}{R_s} + C_p \dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref} - \frac{V_o - V_p}{R_f} \\
C_f \dot{V}_o &= \frac{V_p - V_o}{R_f} + (C_p + C_f) \dot{V}_p + \frac{V_p - V_m}{R_s} - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

resulting in an alternative expression for the same model: (1, 2b, 3b, 4, 5b).

## Three models

This leaves us with the following equations:

\begin{align}
1. && C_m\dot{V}_m = \frac{V_p - V_m}{R_s} - I
\end{align}

\begin{align}
2a. && (C_p + C_f)\dot{V}_p &= \frac{V_o-V_p}{R_f} + \frac{V_m-V_p}{R_s} + C_f\dot{V}_o  + (C_p^* + C_m^*) \dot{V}_\text{ref} \\
2b. && C_f\dot{V}_o &= \frac{V_p-V_o}{R_f} + \frac{V_p-V_m}{R_s} + \left(C_p+C_f\right)\dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

\begin{align}
3a. && \tau_a\dot{V}_o = V_\text{ref} - V_p \\
3b. && \tau_c\dot{V}_p = V_\text{ref} - V_p
\end{align}

\begin{align}
4. && \tau_\text{sum} \dot{V}_\text{ref} = V_c + \alpha R_s^* I_\text{obs} - V_\text{ref}
\end{align}

\begin{align}
5a. && R_f I_\text{obs} = V_o - V_\text{ref} \\
5b. && R_f I_\text{obs} = V_o - V_p
\end{align}

and three models:

1. Sigworth-style **model A** (1, 2a, 3a, 4, 5a)
2. Hybrid-style **model B** (1, 2b, 3b, 4, 5a)
3. Weerakoon/Lei-style **model C** (1, 2b, 3b, 4, 5b)

## Simulations

We now run simulations for a single step from -80 to -20 mV.
The ionic current is set to 0, and we assume all voltages were at -80mV at the start of the step.

```python
import myokit

import numpy as np
import matplotlib.pyplot as plt
```

## Model (1, 2a, 3a, 4, 5a)

We start with a (1, 2a, 3a, 4, 5a) model, with a switch to turn compensations on or off.

```python
mA = myokit.parse_model('''
[[model]]
desc: Compensated model (1, 2a, 3a, 4, 5a)
amp.Vm = -80
amp.Vp = -80
amp.Vo = -80
amp.Vr = -80

[amp]
alpha = 0.7
beta = 1
time = 0 [ms] in [ms] bind time
Vc = -20 [mV] in [mV]
Cm = 20 [pF] in [pF]
Cp = 4.5 [pF] in [pF]
Cf = 0.3 [pF] in [pF]
Cm_est = 20 [pF] in [pF]
Cp_est = 4.5 [pF] in [pF]
Rs = 0.01 [GOhm] in [GOhm]
Rf = 0.025 [GOhm] in [GOhm]
Rs_est = 0.01 [GOhm] in [GOhm]
tau_amp = 50e-6 [ms] in [ms]
tau_sum = 40e-3 [ms] in [ms]
I_inj = (Cp_est + Cm_est) * dot(Vr) * beta
    in [pA]
dot(Vm) = (Vp - Vm) / (Rs * Cm) : Eq 1
    in [mV]
dot(Vp) = ((Vo - Vp) / Rf + (Vm - Vp) / Rs + Cf * dot(Vo) + I_inj) / (Cp + Cf) : Eq 2a
    in [mV]
dot(Vo) = (Vr - Vp) / tau_amp : Eq 3a
    in [mV]
dot(Vr) = (Vc + alpha * Rs_est * I_obs - Vr) / tau_sum : Eq 4
    in [mV]
I_obs = (Vo - Vr) / Rf : Eq 5a
    in [pA]
''')
mA.check_units(myokit.UNIT_STRICT)
```

```python
def plot(m, d, axes=None, label=None):
    if axes is None:
        fig = plt.figure(figsize=(15, 10))
        t_zoom = 0.05
        
        # Top left: Vm
        ax1 = fig.add_subplot(3, 2, 1)
        ax1.set_ylabel('Vm (mV)')
        
        # Top right: Vo
        ax2 = fig.add_subplot(3, 2, 2)
        ax2.set_ylabel('Vo (mV)')
        ins2 = ax2.inset_axes((0.3, 0.35, 0.65, 0.55))
        ins2.set_xlabel('Time (ms)')
        ins2.set_ylabel('Vo (mV)')
        ins2.set_xlim(-0.005, t_zoom)
        ins2.patch.set_alpha(0.5)

        # Middle left: Vp
        ax3 = fig.add_subplot(3, 2, 3)
        ax3.set_xlabel('Time (ms)')
        ax3.set_ylabel('Vp (mV)')
        ins3 = ax3.inset_axes((0.3, 0.15, 0.65, 0.55))
        ins3.set_xlabel('Time (ms)')
        ins3.set_ylabel('Vp (mV)')
        ins3.set_xlim(-0.005, t_zoom)
        ins3.patch.set_alpha(0.5)

        # Middle right: V_ref
        ax4 = fig.add_subplot(3, 2, 4)
        ax4.set_xlabel('Time (ms)')
        ax4.set_ylabel('Vref (mV)')
        ins4 = ax4.inset_axes((0.3, 0.15, 0.65, 0.55))
        ins4.set_xlabel('Time (ms)')
        ins4.set_ylabel('Vref (mV)')
        ins4.set_xlim(-0.005, t_zoom)
        ins4.patch.set_alpha(0.5)
        
        # Lower: I_obs
        ax5 = fig.add_subplot(3, 1, 3)
        ax5.set_xlabel('Time (ms)')
        ax5.set_ylabel('Recorded I (pA)')
        ins5 = ax5.inset_axes((0.3, 0.35, 0.65, 0.55))
        ins5.set_xlabel('Time (ms)')
        ins5.set_ylabel('Recorded I (pA)')
        ins5.set_xlim(-0.001, t_zoom)
        ins5.patch.set_alpha(0.5)

        kw = dict(color='#aaa', ls='--')
        ax1.axhline(m.get('amp.Vm').initial_value().eval(), label='Vm(t=0)', **kw)
        ax1.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
        ax2.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
        ax3.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
        ax4.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
    else:
        [ax1, ax2, ax3, ax4, ax5, ins2, ins3, ins4, ins5] = axes
        
    ax1.plot(d.time(), d['amp.Vm'], label=label)
    if 'amp.Vo' in d:
        ax2.plot(d.time(), d['amp.Vo'])
        ins2.plot(d.time(), d['amp.Vo'])
    ax3.plot(d.time(), d['amp.Vp'])
    ins3.plot(d.time(), d['amp.Vp'])
    ax4.plot(d.time(), d['amp.Vr'])
    ins4.plot(d.time(), d['amp.Vr'])
    ax5.plot(d.time(), d['amp.I_obs'])
    ins5.plot(d.time(), d['amp.I_obs'])
    
    return [ax1, ax2, ax3, ax4, ax5, ins2, ins3, ins4, ins5]
```

```python
sA = myokit.Simulation(mA)
sA.set_constant('amp.alpha', 0)
sA.set_constant('amp.beta', 0)
sA.set_constant('amp.tau_sum', 1e-9)
dA = sA.run(5)
plot(mA, dA)
plt.show()
```

Now we switch compensation on and run again

```python
sA.reset()
sA.set_constant('amp.alpha', 0)
sA.set_constant('amp.beta', 1)
sA.set_constant('amp.tau_sum', 40e-3)
dA = sA.run(5)
print(abs(-20 - dA['amp.Vm'][-1]))
ax = plot(mA, dA)

sA.reset()
sA.set_constant('amp.alpha', 0.4)
dA = sA.run(5)
print(abs(-20 - dA['amp.Vm'][-1]))
ax = plot(mA, dA, axes=ax)

sA.reset()
sA.set_constant('amp.alpha', 0.8)
dA = sA.run(5)
print(abs(-20 - dA['amp.Vm'][-1]))
ax = plot(mA, dA, axes=ax)
```

## Model (1, 2b, 3b, 4, 5a)

\begin{align}
2b. && C_f\dot{V}_o &= \frac{V_p-V_o}{R_f} + \frac{V_p-V_m}{R_s} + \left(C_p+C_f\right)\dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

\begin{align}
3b. && \tau_c\dot{V}_p = V_\text{ref} - V_p
\end{align}

```python
mB = myokit.parse_model('''
[[model]]
desc: Compensated model (1, 2b, 3b, 4, 5a)
amp.Vm = -80
amp.Vp = -80
amp.Vo = -80
amp.Vr = -80

[amp]
alpha = 0.7
beta = 1
time = 0 [ms] in [ms] bind time
Vc = -20 [mV] in [mV]
Cm = 20 [pF] in [pF]
Cp = 4.5 [pF] in [pF]
Cf = 0.3 [pF] in [pF]
Cm_est = 20 [pF] in [pF]
Cp_est = 4.5 [pF] in [pF]
Rs = 0.01 [GOhm] in [GOhm]
Rf = 0.025 [GOhm] in [GOhm]
Rs_est = 0.01 [GOhm] in [GOhm]
tau_c = 50e-6 [ms] * (Cf + Cp) / Cf in [ms]
tau_sum = 40e-3 [ms] in [ms]
I_inj = (Cp_est + Cm_est) * dot(Vr) * beta
    in [pA]
dot(Vm) = (Vp - Vm) / (Rs * Cm) : Eq 1
    in [mV]
dot(Vo) = ((Vp - Vo) / Rf + (Vp - Vm) / Rs + (Cp + Cf) * dot(Vp) - I_inj) / Cf : Eq 2b
    in [mV]
dot(Vp) = (Vr - Vp) / tau_c : Eq 3b
    in [mV]
dot(Vr) = (Vc + alpha * Rs_est * I_obs - Vr) / tau_sum : Eq 4
    in [mV]
I_obs = (Vo - Vr) / Rf : Eq 5a
    in [pA]
''')
mB.check_units(myokit.UNIT_STRICT)
sB = myokit.Simulation(mB)
```

```python
sB.reset()
sB.set_constant('amp.alpha', 0)
sB.set_constant('amp.beta', 1)
sB.set_constant('amp.tau_sum', 40e-3)
dB = sB.run(10)
print(abs(-20 - dB['amp.Vm'][-1]))
ax = plot(mB, dB)

sB.reset()
sB.set_constant('amp.alpha', 0.4)
dB = sB.run(10)
print(abs(-20 - dB['amp.Vm'][-1]))
ax = plot(mB, dB, axes=ax)

sB.reset()
sB.set_constant('amp.alpha', 0.8)
dB = sB.run(10)
print(abs(-20 - dB['amp.Vm'][-1]))
ax = plot(mB, dB, axes=ax)
```

## (1, 2b, 3b, 4, 5b)

\begin{align}
5b. && R_f I_\text{obs} = V_o - V_p
\end{align}

```python
mC = myokit.parse_model('''
[[model]]
desc: Compensated model (1, 2b, 3b, 4, 5b)
amp.Vm = -80
amp.Vp = -80
amp.Vo = -80
amp.Vr = -80

[amp]
alpha = 0.7
beta = 1
time = 0 [ms] in [ms] bind time
Vc = -20 [mV] in [mV]
Cm = 20 [pF] in [pF]
Cp = 4.5 [pF] in [pF]
Cf = 0.3 [pF] in [pF]
Cm_est = 20 [pF] in [pF]
Cp_est = 4.5 [pF] in [pF]
Rs = 0.01 [GOhm] in [GOhm]
Rf = 0.025 [GOhm] in [GOhm]
Rs_est = 0.01 [GOhm] in [GOhm]
tau_c = 50e-6 [ms] * (Cf + Cp) / Cf in [ms]
tau_sum = 40e-3 [ms] in [ms]
I_inj = (Cp_est + Cm_est) * dot(Vr) * beta
    in [pA]
dot(Vm) = (Vp - Vm) / (Rs * Cm) : Eq 1
    in [mV]
dot(Vo) = ((Vp - Vo) / Rf + (Vp - Vm) / Rs + (Cp + Cf) * dot(Vp) - I_inj) / Cf : Eq 2b
    in [mV]
dot(Vp) = (Vr - Vp) / tau_c : Eq 3b
    in [mV]
dot(Vr) = (Vc + alpha * Rs_est * I_obs - Vr) / tau_sum : Eq 4
    in [mV]
I_obs = (Vo - Vp) / Rf : Eq 5b
    in [pA]
''')
mC.check_units(myokit.UNIT_STRICT)
sC = myokit.Simulation(mC)
```

```python
sC.reset()
sC.set_constant('amp.alpha', 0)
sC.set_constant('amp.beta', 1)
sC.set_constant('amp.tau_sum', 40e-3)
dC = sC.run(10)
print(abs(-20 - dC['amp.Vm'][-1]))
ax = plot(mC, dC)

sC.reset()
sC.set_constant('amp.alpha', 0.4)
dC = sC.run(10)
print(abs(-20 - dC['amp.Vm'][-1]))
ax = plot(mC, dC, axes=ax)

sC.reset()
sC.set_constant('amp.alpha', 0.8)
dC = sC.run(10)
print(abs(-20 - dC['amp.Vm'][-1]))
ax = plot(mC, dC, axes=ax)
```

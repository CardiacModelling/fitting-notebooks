# Appendix L: Compensated patch-clamp models

In [Appendix F](./appendix-E-non-ideal-op-amp.ipynb) we compared uncompensated patch-clamp models, omitting voltage offset and leak current for simplicity.
In this appendix, we compare models of _compensated_ patch clamp, again omitting offset and leak.

The schematic is shown below:

<img src="resources/patch-comp-7-no-leak-eoff.png" style="margin:auto" />

## Sigworth-style model (1, 2a, 3a, 4, 5a)

\begin{align}
1. && C_m\dot{V}_m = \frac{V_p - V_m}{R_s} - I
\end{align}

\begin{align}
2a. && (C_p + C_f)\dot{V}_p = \frac{V_o-V_p}{R_f} + C_f\dot{V}_o - \frac{V_p-V_m}{R_s} + (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

\begin{align}
3a. && \tau_a\dot{V}_o = V_\text{ref} - V_p
\end{align}

\begin{align}
4. && \tau_\text{sum} \dot{V}_\text{ref} = V_c + \alpha R_s^* I_\text{obs} - V_\text{ref}
\end{align}

\begin{align}
5a. && R_f I_\text{obs} = V_o - V_\text{ref}
\end{align}

## Alternative op-amp equation (1, 2b, 3b, 4, 5a)

\begin{align}
1. && C_m\dot{V}_m = \frac{V_p - V_m}{R_s} - I
\end{align}

\begin{align}
2b. && C_f\dot{V}_o = \frac{V_p-V_o}{R_f} + \left(C_p+C_f\right)\dot{V}_p + \frac{V_p-V_m}{R_s} - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

\begin{align}
3b. && \tau_c\dot{V}_p = V_\text{ref} - V_p
\end{align}

\begin{align}
4. && \tau_\text{sum} \dot{V}_\text{ref} = V_c + \alpha R_s^* I_\text{obs} - V_\text{ref}
\end{align}

\begin{align}
5a. && R_f I_\text{obs} = V_o - V_\text{ref}
\end{align}

## Lei-style model (1, 2c, 3b, 4, 5a) or (1, 2b, 3b, 4, 5b)

Following Lei et al. (2020), we get

\begin{align}
1. && C_m\dot{V}_m = \frac{V_p - V_m}{R_s} - I && \text{(Equation 2.3)}
\end{align}

\begin{align}
3b. && \tau_c\dot{V}_p = V_c - V_p && \text{(Equation 2.4)}
\end{align}

\begin{align}
4. && \tau_\text{sum} \dot{V}_\text{ref} = V_c + \alpha R_s^* I_\text{obs} - V_\text{ref} && \text{(Equation 2.5)}
\end{align}

and we can combine two equations to find

\begin{align}
I_\text{in} &= I + C_p \dot{V}_p + C_m \dot{V}_m - (C_p^* + C_m^*) \dot{V}_\text{ref} && \text{(Equation 2.6)} \\
            &= \frac{V_p - V_m}{R_s} + C_p \dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref} && \text{(Equation 2.3)}
\end{align}

Then, we define

\begin{align}
5a. && R_f I_\text{obs} = V_o - V_\text{ref} && \text{(No equivalent)}
\end{align}

and insert into Lei et al. Equation 2.7 to find

\begin{align}
R_f C_f \dot{I}_\text{obs} &= I_\text{in} - I_\text{obs} && \text{(Equation 2.7)} \\
C_f (\dot{V}_o - \dot{V}_\text{ref}) &= \frac{V_p - V_m}{R_s} + C_p \dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref} - \frac{V_o - V_\text{ref}}{R_f} &&
\end{align}

\begin{align}
2c. && C_f \dot{V}_o &= \frac{V_\text{ref} - V_o}{R_f} + C_p \dot{V}_p + C_f \dot{V}_\text{ref} + \frac{V_p - V_m}{R_s} - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

For a model (1, 2c, 3b, 4, 5a).

**Alternatively**, and equivalently, we can define

\begin{align}
5b. && R_f I_\text{obs} = V_o - V_p && \text{(No equivalent)}
\end{align}

with which we can derive 2b. from Lei et al. Equation 2.7:

\begin{align}
C_f (\dot{V}_o - \dot{V}_p) &= \frac{V_p - V_m}{R_s} + C_p \dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref} - \frac{V_o - V_p}{R_f} \\
C_f \dot{V}_o &= \frac{V_p - V_o}{R_f} + (C_p + C_f) \dot{V}_p + \frac{V_p - V_m}{R_s} - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

resulting in an alternative expression for the same model: (1, 2b, 3b, 4, 5b).

## Three models

This leaves us with the following equations:

\begin{align}
1. && C_m\dot{V}_m = \frac{V_p - V_m}{R_s} - I
\end{align}

\begin{align}
2a. && (C_p + C_f)\dot{V}_p &= \frac{V_o-V_p}{R_f} + \frac{V_m-V_p}{R_s} + C_f\dot{V}_o  + (C_p^* + C_m^*) \dot{V}_\text{ref} \\
2b. && C_f\dot{V}_o &= \frac{V_p-V_o}{R_f} + \frac{V_p-V_m}{R_s} + \left(C_p+C_f\right)\dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

\begin{align}
3a. && \tau_a\dot{V}_o = V_\text{ref} - V_p \\
3b. && \tau_c\dot{V}_p = V_\text{ref} - V_p
\end{align}

\begin{align}
4. && \tau_\text{sum} \dot{V}_\text{ref} = V_c + \alpha R_s^* I_\text{obs} - V_\text{ref}
\end{align}

\begin{align}
5a. && R_f I_\text{obs} = V_o - V_\text{ref} \\
5b. && R_f I_\text{obs} = V_o - V_p
\end{align}

and three models:

1. Sigworth-style **model A** (1, 2a, 3a, 4, 5a)
2. Hybrid-style **model B** (1, 2b, 3b, 4, 5a)
3. Weerakoon/Lei-style **model C** (1, 2b, 3b, 4, 5b)

## Simulations

We now run simulations for a single step from -80 to -20 mV.
The ionic current is set to 0, and we assume all voltages were at -80mV at the start of the step.

```python
import myokit

import numpy as np
import matplotlib.pyplot as plt
```

## Sigworth-style model (1, 2a, 3a, 4, 5a)

We start with a (1, 2a, 3a, 4, 5a) model, with a switch to turn compensations on or off.

```python
mA = myokit.parse_model('''
[[model]]
desc: Compensated model (1, 2a, 3a, 4, 5a)
amp.Vm = -80
amp.Vp = -80
amp.Vo = -80
amp.Vr = -80

[amp]
alpha = 0.7
beta = 1
time = 0 [ms] in [ms] bind time
Vc = -20 [mV] in [mV]
Cm = 20 [pF] in [pF]
Cp = 4.5 [pF] in [pF]
Cf = 0.3 [pF] in [pF]
Cm_est = 20 [pF] in [pF]
Cp_est = 4.5 [pF] in [pF]
Rs = 0.01 [GOhm] in [GOhm]
Rf = 0.025 [GOhm] in [GOhm]
Rs_est = 0.01 [GOhm] in [GOhm]
tau_amp = 50e-6 [ms] in [ms]
tau_sum = 40e-3 [ms] in [ms]
I_inj = (Cp_est + Cm_est) * dot(Vr) * beta
    in [pA]
dot(Vm) = (Vp - Vm) / (Rs * Cm) : Eq 1
    in [mV]
dot(Vp) = ((Vo - Vp) / Rf + (Vm - Vp) / Rs + Cf * dot(Vo) + I_inj) / (Cp + Cf) : Eq 2a
    in [mV]
dot(Vo) = (Vr - Vp) / tau_amp : Eq 3a
    in [mV]
dot(Vr) = (Vc + alpha * Rs_est * I_obs - Vr) / tau_sum : Eq 4
    in [mV]
I_obs = (Vo - Vr) / Rf : Eq 5a
    in [pA]
''')
mA.check_units(myokit.UNIT_STRICT)
```

```python
def plot(m, d, axes=None, label=None, ls=None, altins=False):
    if axes is None:
        fig = plt.figure(figsize=(15, 10))
        t_zoom = 0.05
        
        # Top left: Vm
        ax1 = fig.add_subplot(3, 2, 1)
        ax1.set_ylabel('Vm (mV)')
        
        # Top right: Vo
        ax2 = fig.add_subplot(3, 2, 2)
        ax2.set_ylabel('Vo (mV)')
        ins2 = ax2.inset_axes((0.3, 0.35 if altins else 0.20, 0.65, 0.55))
        ins2.set_xlabel('Time (ms)')
        ins2.set_ylabel('Vo (mV)')
        ins2.set_xlim(-0.005, t_zoom)
        ins2.patch.set_alpha(0.5)

        # Middle left: Vp
        ax3 = fig.add_subplot(3, 2, 3)
        ax3.set_xlabel('Time (ms)')
        ax3.set_ylabel('Vp (mV)')
        ins3 = ax3.inset_axes((0.3, 0.20, 0.65, 0.55))
        ins3.set_xlabel('Time (ms)')
        ins3.set_ylabel('Vp (mV)')
        ins3.set_xlim(-0.005, t_zoom)
        ins3.patch.set_alpha(0.5)

        # Middle right: V_ref
        ax4 = fig.add_subplot(3, 2, 4)
        ax4.set_xlabel('Time (ms)')
        ax4.set_ylabel('Vref (mV)')
        ins4 = ax4.inset_axes((0.3, 0.20, 0.65, 0.55))
        ins4.set_xlabel('Time (ms)')
        ins4.set_ylabel('Vref (mV)')
        ins4.set_xlim(-0.005, t_zoom)
        ins4.patch.set_alpha(0.5)
        
        # Lower: I_obs
        ax5 = fig.add_subplot(3, 1, 3)
        ax5.set_xlabel('Time (ms)')
        ax5.set_ylabel('Recorded I (pA)')
        ins5 = ax5.inset_axes((0.3, 0.35 if altins else 0.20, 0.65, 0.55))
        ins5.set_xlabel('Time (ms)')
        ins5.set_ylabel('Recorded I (pA)')
        ins5.set_xlim(-0.001, t_zoom)
        ins5.patch.set_alpha(0.5)

        kw = dict(color='#aaa', ls='--')
        ax1.axhline(m.get('amp.Vm').initial_value().eval(), label='Vm(t=0)', **kw)
        ax1.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
        ax2.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
        ax3.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
        ax4.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
    else:
        [ax1, ax2, ax3, ax4, ax5, ins2, ins3, ins4, ins5] = axes
        
    ax1.plot(d.time(), d['amp.Vm'], label=label, ls=ls)
    if 'amp.Vo' in d:
        ax2.plot(d.time(), d['amp.Vo'], ls=ls)
        ins2.plot(d.time(), d['amp.Vo'], ls=ls)
    ax3.plot(d.time(), d['amp.Vp'], ls=ls)
    ins3.plot(d.time(), d['amp.Vp'], ls=ls)
    ax4.plot(d.time(), d['amp.Vr'], ls=ls)
    ins4.plot(d.time(), d['amp.Vr'], ls=ls)
    ax5.plot(d.time(), d['amp.I_obs'], ls=ls)
    ins5.plot(d.time(), d['amp.I_obs'], ls=ls)
    
    return [ax1, ax2, ax3, ax4, ax5, ins2, ins3, ins4, ins5]
```

```python
sA = myokit.Simulation(mA)
sA.set_constant('amp.alpha', 0)
sA.set_constant('amp.beta', 0)
sA.set_constant('amp.tau_sum', 1e-9)
dA = sA.run(5)
ax = plot(mA, dA, altins=True)
plt.show()
```

Now we switch compensation on and run again

```python
sA.reset()
sA.set_constant('amp.alpha', 0)
sA.set_constant('amp.beta', 1)
sA.set_constant('amp.tau_sum', 40e-3)
dA = sA.run(5)
print(abs(-20 - dA['amp.Vm'][-1]))
ax = plot(mA, dA, label='$\\alpha$ = 0')

sA.reset()
sA.set_constant('amp.alpha', 0.4)
dA = sA.run(5)
print(abs(-20 - dA['amp.Vm'][-1]))
ax = plot(mA, dA, axes=ax, label='$\\alpha$ = 0.4')

sA.reset()
sA.set_constant('amp.alpha', 0.8)
dA = sA.run(5)
print(abs(-20 - dA['amp.Vm'][-1]))
ax = plot(mA, dA, axes=ax, label='$\\alpha$ = 0.8')
ax[0].legend()
plt.show()
```

## Alternative op-amp equation model (1, 2b, 3b, 4, 5a)

\begin{align}
2b. && C_f\dot{V}_o &= \frac{V_p-V_o}{R_f} + \frac{V_p-V_m}{R_s} + \left(C_p+C_f\right)\dot{V}_p - (C_p^* + C_m^*) \dot{V}_\text{ref}
\end{align}

\begin{align}
3b. && \tau_c\dot{V}_p = V_\text{ref} - V_p
\end{align}

```python
mB = myokit.parse_model('''
[[model]]
desc: Compensated model (1, 2b, 3b, 4, 5a)
amp.Vm = -80
amp.Vp = -80
amp.Vo = -80
amp.Vr = -80

[amp]
alpha = 0.7
beta = 1
time = 0 [ms] in [ms] bind time
Vc = -20 [mV] in [mV]
Cm = 20 [pF] in [pF]
Cp = 4.5 [pF] in [pF]
Cf = 0.3 [pF] in [pF]
Cm_est = 20 [pF] in [pF]
Cp_est = 4.5 [pF] in [pF]
Rs = 0.01 [GOhm] in [GOhm]
Rf = 0.025 [GOhm] in [GOhm]
Rs_est = 0.01 [GOhm] in [GOhm]
tau_c = 50e-6 [ms] * (Cf + Cp) / Cf in [ms]
tau_sum = 40e-3 [ms] in [ms]
I_inj = (Cp_est + Cm_est) * dot(Vr) * beta
    in [pA]
dot(Vm) = (Vp - Vm) / (Rs * Cm) : Eq 1
    in [mV]
dot(Vo) = ((Vp - Vo) / Rf + (Vp - Vm) / Rs + (Cp + Cf) * dot(Vp) - I_inj) / Cf : Eq 2b
    in [mV]
dot(Vp) = (Vr - Vp) / tau_c : Eq 3b
    in [mV]
dot(Vr) = (Vc + alpha * Rs_est * I_obs - Vr) / tau_sum : Eq 4
    in [mV]
I_obs = (Vo - Vr) / Rf : Eq 5a
    in [pA]
''')
mB.check_units(myokit.UNIT_STRICT)
sB = myokit.Simulation(mB)

print(f'tau_c = {mB.get("amp.tau_c").eval()} ms')
print(f'tau_c = {mB.get("amp.tau_c").eval() * 1e3} us')
```

```python
sB.reset()
sB.set_constant('amp.alpha', 0)
dB = sB.run(5)
print(abs(-20 - dB['amp.Vm'][-1]))
ax = plot(mB, dB, label='$\\alpha$ = 0')

sB.reset()
sB.set_constant('amp.alpha', 0.4)
dB = sB.run(5)
print(abs(-20 - dB['amp.Vm'][-1]))
ax = plot(mB, dB, axes=ax, label='$\\alpha$ = 0.4')

sB.reset()
sB.set_constant('amp.alpha', 0.8)
dB = sB.run(5)
print(abs(-20 - dB['amp.Vm'][-1]))
ax = plot(mB, dB, axes=ax, label='$\\alpha$ = 0.8')

ax[0].legend()
plt.show()
```

## Weerakoon/Lei-style model (1, 2b, 3b, 4, 5b)

\begin{align}
5b. && R_f I_\text{obs} = V_o - V_p
\end{align}

```python
mC = myokit.parse_model('''
[[model]]
desc: Compensated model (1, 2b, 3b, 4, 5b)
amp.Vm = -80
amp.Vp = -80
amp.Vo = -80
amp.Vr = -80

[amp]
alpha = 0.7
beta = 1
time = 0 [ms] in [ms] bind time
Vc = -20 [mV] in [mV]
Cm = 20 [pF] in [pF]
Cp = 4.5 [pF] in [pF]
Cf = 0.3 [pF] in [pF]
Cm_est = 20 [pF] in [pF]
Cp_est = 4.5 [pF] in [pF]
Rs = 0.01 [GOhm] in [GOhm]
Rf = 0.025 [GOhm] in [GOhm]
Rs_est = 0.01 [GOhm] in [GOhm]
tau_c = 50e-6 [ms] * (Cf + Cp) / Cf in [ms]
tau_sum = 40e-3 [ms] in [ms]
I_inj = (Cp_est + Cm_est) * dot(Vr) * beta
    in [pA]
dot(Vm) = (Vp - Vm) / (Rs * Cm) : Eq 1
    in [mV]
dot(Vo) = ((Vp - Vo) / Rf + (Vp - Vm) / Rs + (Cp + Cf) * dot(Vp) - I_inj) / Cf : Eq 2b
    in [mV]
dot(Vp) = (Vr - Vp) / tau_c : Eq 3b
    in [mV]
dot(Vr) = (Vc + alpha * Rs_est * I_obs - Vr) / tau_sum : Eq 4
    in [mV]
I_obs = (Vo - Vp) / Rf : Eq 5b
    in [pA]
''')
mC.check_units(myokit.UNIT_STRICT)
sC = myokit.Simulation(mC)
```

```python
sC.reset()
sC.set_constant('amp.alpha', 0)
dC = sC.run(5)
print(abs(-20 - dC['amp.Vm'][-1]))
ax = plot(mC, dC, label='$\\alpha$ = 0')

sC.reset()
sC.set_constant('amp.alpha', 0.4)
dC = sC.run(5)
print(abs(-20 - dC['amp.Vm'][-1]))
ax = plot(mC, dC, axes=ax, label='$\\alpha$ = 0.4')

sC.reset()
sC.set_constant('amp.alpha', 0.8)
dC = sC.run(5)
print(abs(-20 - dC['amp.Vm'][-1]))
ax = plot(mC, dC, axes=ax, label='$\\alpha$ = 0.8')

ax[0].legend()
plt.show()
```

## Original Lei formulation

Just to check

```python
mD = myokit.parse_model('''
[[model]]
desc: Lei formulation. Should equal (1, 2b, 3b, 4, 5b)
amp.Vm = -80
amp.Vp = -80
amp.Vr = -80
amp.I_obs = 0

[amp]
alpha = 0.7
beta = 1
time = 0 [ms] in [ms] bind time
Vc = -20 [mV] in [mV]
Cm = 20 [pF] in [pF]
Cp = 4.5 [pF] in [pF]
Cf = 0.3 [pF] in [pF]
Cm_est = 20 [pF] in [pF]
Cp_est = 4.5 [pF] in [pF]
Rs = 0.01 [GOhm] in [GOhm]
Rf = 0.025 [GOhm] in [GOhm]
Rs_est = 0.01 [GOhm] in [GOhm]
tau_c = 50e-6 [ms] * (Cf + Cp) / Cf in [ms]
tau_sum = 40e-3 [ms] in [ms]
dot(Vm) = (Vp - Vm) / (Rs * Cm) : (2.3) = Eq 1
    in [mV]
dot(Vp) = (Vr - Vp) / tau_c : (2.4) = Eq 3b
    in [mV]
dot(Vr) = (Vc + alpha * Rs_est * I_obs - Vr) / tau_sum : (2.5) = Eq 4
    in [mV]
I_in = Cp * dot(Vp) + Cm * dot(Vm) - (Cp_est + Cm_est) * dot(Vr) * beta : (2.6)
    in [pA]
dot(I_obs) = (I_in - I_obs) / (Rf * Cf) : (2.7)
    in [pA]
''')
mD.check_units(myokit.UNIT_STRICT)
sD = myokit.Simulation(mD)
```

```python
sD.reset()
sD.set_constant('amp.alpha', 0)
dD = sD.run(5)
print(abs(-20 - dD['amp.Vm'][-1]))
ax = plot(mD, dD, label='$\\alpha$ = 0')

sD.reset()
sD.set_constant('amp.alpha', 0.4)
dD = sD.run(5)
print(abs(-20 - dD['amp.Vm'][-1]))
ax = plot(mD, dD, axes=ax, label='$\\alpha$ = 0.4')

sD.reset()
sD.set_constant('amp.alpha', 0.8)
dD = sD.run(5)
print(abs(-20 - dD['amp.Vm'][-1]))
ax = plot(mD, dD, axes=ax, label='$\\alpha$ = 0.8')

ax[0].legend()
plt.show()
```

## Model predictions overlayed

```python
for s in (sA, sB, sC, sD):
    s.reset()
    s.set_constant('amp.alpha', 0.8)
    s.set_tolerance(1e-10, 1e-10)

t = 2
dt = t * 1e-4
dA = sA.run(t, log_interval=dt).npview()
dB = sB.run(t, log_interval=dt).npview()
dC = sC.run(t, log_interval=dt).npview()
dD = sD.run(t, log_interval=dt).npview()

ax = plot(mA, dA, label='A')
plot(mB, dB, ax, label='B', ls='--')
plot(mC, dC, ax, label='C', ls='-.')
plot(mD, dD, ax, label='D', ls=':')
ax[0].legend()
plt.show()
```

```python
fig = plt.figure(figsize=(15, 15))
axes = []
names = ('A', 'B', 'C', 'D')
logs = (dA, dB, dC, dD)
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']
ylim0 = {'amp.I_obs': (-6900, 1900)}
ylim1 = {'amp.I_obs': (-10, 10)}
var = 'amp.I_obs'

for i in range(4):
    for j in range(4):
        if i == j:
            continue
        ax = fig.add_subplot(4, 4, 4 * i + j + 1)        
        if i == 3 or i == j - 1:
            ax.set_xlabel(names[j])
        else:
            ax.set_xticklabels([])
        if j == 0 or j == i + 1:
            ax.set_ylabel(names[i])
        else:
            ax.set_yticklabels([])
        
        if i > j:
            ax.plot(logs[j].time(), logs[j][var], color=colors[j], label=f'{names[j]} {var}')
            ax.plot(logs[i].time(), logs[i][var], color=colors[i], label=f'{names[i]} {var}', ls='--')            
            if var in  ylim0:
                ax.set_ylim(*ylim0[var])
        else:
            ax.plot(logs[i].time(), logs[i][var] - logs[j][var], 
                    color='k', label=f'{names[i]} - {names[j]}')
            if var in  ylim1:
                ax.set_ylim(*ylim1[var])
        ax.legend()

plt.show()
```

This shows that:

1. A, B, and C give several results (for these parameters),
2. A looks to be the odd one out, with B and C looking slightly more similar to each other,
3. As shown, C is a reformulation of the Lei et al. compensated model.

```python
def plot(m, d, axes=None, label=None, ls=None, altins=False):
    if axes is None:
        fig = plt.figure(figsize=(15, 10))
        t_zoom = 0.05
        
        # Top left: Vm
        ax1 = fig.add_subplot(3, 2, 1)
        ax1.set_ylabel('Vm (mV)')
        
        # Top right: Vo
        ax2 = fig.add_subplot(3, 2, 2)
        ax2.set_ylabel('Vo (mV)')
        ins2 = ax2.inset_axes((0.3, 0.35 if altins else 0.20, 0.65, 0.55))
        ins2.set_xlabel('Time (ms)')
        ins2.set_ylabel('Vo (mV)')
        ins2.set_xlim(-0.005, t_zoom)
        ins2.patch.set_alpha(0.5)

        # Middle left: Vp
        ax3 = fig.add_subplot(3, 2, 3)
        ax3.set_xlabel('Time (ms)')
        ax3.set_ylabel('Vp (mV)')
        ins3 = ax3.inset_axes((0.3, 0.20, 0.65, 0.55))
        ins3.set_xlabel('Time (ms)')
        ins3.set_ylabel('Vp (mV)')
        ins3.set_xlim(-0.005, t_zoom)
        ins3.patch.set_alpha(0.5)

        # Middle right: V_ref
        ax4 = fig.add_subplot(3, 2, 4)
        ax4.set_xlabel('Time (ms)')
        ax4.set_ylabel('Vref (mV)')
        ins4 = ax4.inset_axes((0.3, 0.20, 0.65, 0.55))
        ins4.set_xlabel('Time (ms)')
        ins4.set_ylabel('Vref (mV)')
        ins4.set_xlim(-0.005, t_zoom)
        ins4.patch.set_alpha(0.5)
        
        # Lower: I_obs
        ax5 = fig.add_subplot(3, 1, 3)
        ax5.set_xlabel('Time (ms)')
        ax5.set_ylabel('Recorded I (pA)')
        ins5 = ax5.inset_axes((0.3, 0.35 if altins else 0.20, 0.65, 0.55))
        ins5.set_xlabel('Time (ms)')
        ins5.set_ylabel('Recorded I (pA)')
        ins5.set_xlim(-0.001, t_zoom)
        ins5.patch.set_alpha(0.5)

        kw = dict(color='#aaa', ls='--')
        ax1.axhline(m.get('amp.Vm').initial_value().eval(), label='Vm(t=0)', **kw)
        ax1.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
        ax2.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
        ax3.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
        ax4.axhline(m.get('amp.Vc').eval(), label='Vc', **kw)
    else:
        [ax1, ax2, ax3, ax4, ax5, ins2, ins3, ins4, ins5] = axes
        
    ax1.plot(d.time(), d['amp.Vm'], label=label, ls=ls)
    if 'amp.Vo' in d:
        ax2.plot(d.time(), d['amp.Vo'], ls=ls)
        ins2.plot(d.time(), d['amp.Vo'], ls=ls)
    ax3.plot(d.time(), d['amp.Vp'], ls=ls)
    ins3.plot(d.time(), d['amp.Vp'], ls=ls)
    ax4.plot(d.time(), d['amp.Vr'], ls=ls)
    ins4.plot(d.time(), d['amp.Vr'], ls=ls)
    ax5.plot(d.time(), d['amp.I_obs'], ls=ls)
    ins5.plot(d.time(), d['amp.I_obs'], ls=ls)
    
    return [ax1, ax2, ax3, ax4, ax5, ins2, ins3, ins4, ins5]
```

```python
mA = myokit.parse_model('''
[[model]]
desc: Compensated model (1, 2a, 3a, 4, 5a)
amp.Vm = -80
amp.Vp = -80
amp.Vo = -80
amp.Vr = -80

[amp]
alpha = 0.7
beta = 1
time = 0 [ms] in [ms] bind time
Vc = -20 [mV] in [mV]
Cm = 20 [pF] in [pF]
Cp = 4.5 [pF] in [pF]
Cf = 0.3 [pF] in [pF]
Cm_est = 20 [pF] in [pF]
Cp_est = 4.5 [pF] in [pF]
Rs = 0.01 [GOhm] in [GOhm]
Rf = 0.025 [GOhm] in [GOhm]
Rs_est = 0.01 [GOhm] in [GOhm]
tau_amp = 50e-6 [ms] in [ms]
tau_sum = 40e-3 [ms] in [ms]
I_inj = (Cp_est + Cm_est) * dot(Vr) * beta
    in [pA]
dot(Vm) = (Vp - Vm) / (Rs * Cm) : Eq 1
    in [mV]
dot(Vp) = ((Vo - Vp) / Rf + (Vm - Vp) / Rs + Cf * dot(Vo) + I_inj) / (Cp + Cf) : Eq 2a
    in [mV]
dot(Vo) = (Vr - Vp) / tau_amp : Eq 3a
    in [mV]
dot(Vr) = (Vc + alpha * Rs_est * I_obs - Vr) / tau_sum : Eq 4
    in [mV]
I_obs = (Vo - Vr) / Rf : Eq 5a
    in [pA]
''')
mA.check_units(myokit.UNIT_STRICT)
```

```python
sA = myokit.Simulation(mA)
dA = sA.run(5)
ax = plot(mA, dA)
plt.show()
```

```python
sA.reset()
sA.set_constant('amp.tau_amp', 1e-10)
dB = sA.run(5)
ax = plot(mA, dA)
ax = plot(mA, dB, axes=ax)
plt.show()
```

```python
sA.reset()
sA.set_constant('amp.tau_sum', 1e-10)
dC = sA.run(5)
ax = plot(mA, dA)
ax = plot(mA, dB, axes=ax)
ax = plot(mA, dC, axes=ax)
plt.show()
```

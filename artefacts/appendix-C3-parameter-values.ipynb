### Origin and definition

Consists _at least_ of the pipette resistance, followed by the resistance of the connection to the cell.
Many of the papers in the [references](./references.ipynb) page have schematics.

The Axon pClamp manual 10.4.1 defines "access resistance" $R_a$ as
> Access resistance ($R_a$) is the sum of the electrode resistance and resistance due to current-impeding factors near the electrode tip, e.g. cellular debris, etc.
> ...
> Electrode resistance ($R_e$), also called pipette resistance ($R_p$), is the resistance due to the electrode.
> It does not include resistance due to environmental current-impeding factors near the electrode tip, e.g. cellular debris, air bubbles, poorly conducting solution etc.
> ...
> $R_a = R_e + R_\text{debris}$

#### Pipette resistance

The pipette resistance is visible before the seal is made, and indicates the size of the aperture in the pipette (plus an offset for whatever resistance was incurred before the tip).
Once a seal is made, it adds to the total series/access resistance.

### Origin and definition

Depends on the cell size.
Usually taken to be about 1pF per $\text{cm}^2$, although when error bars are given it can easily be 2.

The membrane of ventricular cells is often estimated by approximating it as a cylinder and then multiplying by 2 to account for the membrane's many tubules and invaginations.

### Origin and definition

More properly called the "fast" or "parasitic" capacitance, it represents the lumped capacitances of the pipette and many things attached to the pipette.

The "fast capacitance" $C_p$ is described by Sigworth (1995) as

> ...$C_p$ is the capacitance of the pipette and holder...
> ...fast capacitance $C_p$, which arises mainly in the pipette and holder and has negligible series resistance...
> ...the capacitance $C_p$, which includes contributions from the input connector, the pipette
holder, and the capacitance of the pipette itself.

Poler et al. (2005) blame it mostly on other sources:

> The deeper the microelectrode is immersed, the higher the capacitance (typically 1 pF/mm). In addition, the pipette holder also introduces a capacity of 1–2 pF, and the input amplifier adds another 2–10 pF, resulting in a total stray capacitance of 10–15 pF

The EPC-10 manual mentions that, _without anything attached to the probe input_, it measures about 1-1.5pF.

#### Two-component model and lag

Compensation circuitry often splits $C_p$ into two components, and provides an additional "lag" control for its filtering.

The Axopatch 200b splits $C_p$ compensation into fast (0-10pF, 0.2-2us lag) and slow (0-1pF, 0.1-10ms lag).
The Axopatch 1D front panel displays fast (0-10pF, 0.2-5us lag) and slow (0-1pF, 0.1-10ms lag).
The EPC-9 uses a tau from 0.5 to 8 us.
The PC-505B has fast (0-15pF, 0.33-8.5us lag) and very fast (0-5pF, 0.1-1.75us).

Names are typically of headstages rather than amplifiers.

### Origin and definition

The voltage drop over $R_f$ is used to measure currents. To measure small currents, we make it large.
The capacitance $C_f$ is either incurred accidentally, or added by design (Weerakoon 2009).

####  $\tau_f$ might not be the relevant value for HEKA
The HEKA values are presumably compensated as described in [Sigworth 1995a](https://doi.org/10.1007/978-1-4419-1229-9_4), section 2.3, which is said to reduce the amplifier's pole at $\tau_1 \approx \tau_f$ from $\tau_1 \approx \tau_f \approx 1\text{ms}$ to an effective value of 4 $\mu$s.

#### 495 MOhm ?
[Sigworth 1995b](https://doi.org/10.1016/0165-0270(94)00128-4) shows that, in the EPC-9, the 50GOhm resistor is always connected, resulting in the 495MOhm for default gain instead of 500MOhm.
This is presumably true for the EPC10 too, but I can't find a reference for that.

#### $C_f$ is a simplification

- [Finkel 1991](https://www.amazon.co.uk/Molecular-Neurobiology-Practical-Approach/dp/0199631093) notes that "In practice, $C_f$ consists of many resistor-capacitor components that make the frequency response of $R_f$ very complicated".
- [Sigworth 1995a](https://doi.org/10.1007/978-1-4419-1229-9_4) discusses scenarios where $C_f$ is evenly distributed along the resistor, and notes that this need not be the case.
- [Sigworth 1995b](https://doi.org/10.1016/0165-0270(94)00128-4) states that "both the resistance of the 50 GOhm resistor and its stray capacitance can vary considerably" (presumably between components with identical specs), so that compensation circuitry needs to be adjustable.

#### Capacitor-feedback amplifiers

Some amplifiers (e.g. the Axon 200B in "single channel mode" - but not in whole-cell mode) use a "capacitor-feedback" system, where the resistor $R_f$ is omitted entirely.
This requires some special tricks that are not covered in these notebooks.
For more on this, see [Sigworth 1995a](https://doi.org/10.1007/978-1-4419-1229-9_4), [Levis & Rae 1992](https://doi.org/10.1016/0076-6879(92)07004-8), or [Finkel 1991](https://www.amazon.co.uk/Molecular-Neurobiology-Practical-Approach/dp/0199631093).
All three are book chapters but in my cupboard if you need them.

The HEKA manuals describe $10 \mu s$ as medium, while on the Axopatches it is the value at 12 o'clock of the potentiometer. It seems safe to assume this is a good default value.

### Origin and definition

This is a deliberately added 1st order filter over $V_c$, used to reduce fast capacitance artefacts (by "rounding") and make series resistance compensation more stable.
